Calcul de la moyenne en excluant les valeurs aberrantes et remplacement des valeurs aberrantes par la moyenne.

```python
X= [5,10,12,13,12,19]
```

```python
import pandas as pnd
df = pnd.DataFrame(X,columns=(["note"]))
```

```python
#Calcul des moustaches min et max

import numpy as np

Q1 = np.quantile(X,0.25)
print("Q1:",Q1)
Q3 = np.quantile(X,0.75)
print("Q3:",Q3)

EIQ = Q3-Q1
moustacheMin = Q1-1.5*EIQ
moustacheMax = Q3+1.5*EIQ

print("Moustache min:",moustacheMin)
print("Moustache max:",moustacheMax)
```

```python
#Selection des données à garder
A_Garder = []
for index, row in df.iterrows():
    if (row["note"] >= moustacheMin and row["note"]<=moustacheMax):
      A_Garder.append(row["note"])
print("Donnees à garder:",A_Garder)


#Calcul de la moyenne
moyenne = np.mean(A_Garder)
print("Moyenne :",moyenne)
```

```python
for index, row in df.iterrows():
    if (row["note"] < moustacheMin or row["note"]>moustacheMax):
      
      #On remplace la valeur par la moyenne en se basant sur son index
      df.loc[index,"note"] = moyenne
```
